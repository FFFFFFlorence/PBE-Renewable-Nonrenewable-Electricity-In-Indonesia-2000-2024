```python
import pandas as pd
print(pd.__version__)
import glob
import os
```

```python
folder_path = r"C:\Florence\Campus stuff\Semester 5\Pemodelan Bidang Energi - PBE\TUBES - 11\Data Storage\Non-Renewable Data\CSVs"

files = [
    f for f in glob.glob(os.path.join(folder_path, "*.csv"))
    if not f.endswith("elec-fossil-nuclear-renewables.csv")
]

df_other = pd.concat(
    (pd.read_csv(f) for f in files),
    ignore_index=True
)

df_nuclear = pd.read_csv(
    os.path.join(folder_path, "elec-fossil-nuclear-renewables.csv")
)
    
files
```

```python
df_other.columns
```

```python
df_other_sort = df_other[
    (df_other["Entity"] == "Indonesia") &
    (df_other["Year"] >= 2000) &
    (df_other["Year"] <= 2024)
]

df_other_merge = (
    df_other_sort
    .groupby(["Entity", "Year"], as_index=False)
    .first()
)
```

```python
df_other_merge.head()
```

```python
df_nuclear = df_nuclear[
    [
        "Entity",
        "Year",
        "Electricity from nuclear - TWh (adapted for visualization of chart elec-fossil-nuclear-renewables)"
    ]
]

df_nuclear = df_nuclear.rename(
    columns={"Electricity from nuclear - TWh (adapted for visualization of chart elec-fossil-nuclear-renewables)": "Electricity from nuclear - TWh"}
)
```

```python
df_nuclear = df_nuclear[
    (df_nuclear["Entity"] == "Indonesia") &
    (df_nuclear["Year"].between(2000, 2024))
]
```

```python
df_final = pd.merge(
    df_other_merge,
    df_nuclear,
    on=["Entity", "Year"],
    how="left"
)
```

```python
df_final.head()
```

```python
df_final.to_csv("non-renewable-electricity-Indonesia_2000_2024.csv", index=False)
```
