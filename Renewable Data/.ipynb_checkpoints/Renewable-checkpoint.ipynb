```python
import pandas as pd
print(pd.__version__)
import glob
import os
```

```python
folder_path = r"C:\Florence\Campus stuff\Semester 5\Pemodelan Bidang Energi - PBE\TUBES - 11\Data Storage\Renewable Data\CSVs"
files = [
    f for f in glob.glob(os.path.join(folder_path, "*.csv"))
    if not f.endswith("installed-geothermal-capacity.csv")
]

df_other = pd.concat(
    (pd.read_csv(f) for f in files), 
    ignore_index=True
)

df_gthermal = pd.read_csv(
    os.path.join(folder_path, "installed-geothermal-capacity.csv")
)

files
```

```python
df_other.columns
```

```python
df_other_sort = df_other[
    (df_other["Entity"] == "Indonesia") &
    (df_other["Year"] >= 2000) &
    (df_other["Year"] <= 2024)
]

df_other_merge = (
    df_other_sort
    .groupby(["Entity", "Year"], as_index=False)
    .first()
)

df_other_merge = df_other_merge.rename(
    columns={"Biofuels production - TWh": "Electricity from biofuels - TWh"}
)
```

```python
df_other_merge.head()
```

```python
df_gthermal = df_gthermal[
    [
        "Entity",
        "Year",
        "Geothermal capacity - TWh"
    ]
]

df_gthermal = df_gthermal.rename(
    columns={"Geothermal capacity - TWh": "Electricity from geothermal - TWh"}
)
```

```python
df_gthermal = df_gthermal[
    (df_gthermal["Entity"] == "Indonesia") &
    (df_gthermal["Year"] >= 2000) &
    (df_gthermal["Year"] <= 2024)
]
```

```python
df_final = pd.merge(
    df_other_merge,
    df_gthermal,
    on=["Entity", "Year"],
    how="left"
)
```

```python
df_final.head()
```

```python
df_final.to_csv("renewable-electricity-Indonesia_2000_2024.csv", index=False)
```
